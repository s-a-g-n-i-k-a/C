```python
################################importing dataframes,libraries and modules##################################

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score
```

```python
from sklearn import linear_model
from sklearn.linear_model import LogisticRegression
from sklearn import tree
```

```python
df=pd.read_csv("canada_per_capita_income (1).csv")
df
```

```python

#######################train_test_split#############################
input=df.drop("per capita income (US$)",axis='columns')
output=df["per capita income (US$)"]
x_train,x_test,y_train,y_test=train_test_split(df[['year']],df["per capita income (US$)"],train_size=0.7, test_size=0.3)
```

```python
linear_reg_model=linear_model.LinearRegression()
linear_reg_model.fit(x_train,y_train)
linear_reg_model.predict(x_test)
linear_score=linear_reg_model.score(x_test,y_test)
print(linear_score)
# accuracy_score=accuracy_score(x_test,y_test)
# accuracy_score ################test_accuracy=0.8899075778647251
```

```python
decision_tree=tree.DecisionTreeRegressor()
decision_tree.fit(x_train,y_train)
decision_tree.predict(x_test)
print(decision_tree.score(x_test,y_test))
# accuracy_score=accuracy_score(y_train,y_test)
# accuracy_score ############test_accuracy=0.9659159016658799
```

```python
import seaborn as sns
```

```python
%matplotlib inline
plt.scatter(df.year,df["per capita income (US$)"],color='red',marker='+')

# regression Plot
plt.figure(figsize=(10, 6))
sns.regplot(x='year', y='per capita income (US$)', data=df, marker='+', color='red', scatter_kws={'s': 100})
plt.xlabel('year')
plt.ylabel('per capita income (US$)')
plt.title('Scatter Plot with Regression Line')
plt.grid(True)
plt.tight_layout()
plt.show()


# decision_tree_Plot
plt.figure(figsize=(10, 6))
plt.scatter(df.year,df["per capita income (US$)"],color='red',marker='+')
plt.xlabel('year')
plt.ylabel('per capita income (US$)')
plt.plot(x_test, y_test, color='blue', linewidth=2, label='Decision Tree Prediction')
plt.title('Decision Tree Regression')
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```
