```python
##############################################################################crosss validation #############################################################################
import pandas as pd
from sklearn.datasets import load_digits
digits = load_digits()
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
dir(digits)
df=pd.DataFrame(digits.data,columns=digits.feature_names)
df["target"]=digits.target
df
x=df.drop(['target'],axis="columns")
y=df.target
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
```

K-fold Cross Validation Methods\\
1. use all the available data for training the model\\
2. split the model into 80-20% training and testing dataset\\
3. use multiple folds and then averaging them out => cross fold

```python
lr=LogisticRegression(max_iter=100)
lr.fit(x_train,y_train)
lr.score(x_test,y_test)
```

```python
svm=SVC(kernel='linear',random_state=0)
svm.fit(x_train,y_train)
svm.score(x_test,y_test)
```

```python
random_forest=RandomForestClassifier(n_estimators=40)
random_forest.fit(x_train,y_train)
random_forest.score(x_test,y_test)
```

```python
def get_score(model,x_train,x_test,y_train,y_test):
  model.fit(x_train,y_train)
  return model.score(x_test,y_test)
```
