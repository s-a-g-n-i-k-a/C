```python
import pandas as pd
from sklearn.datasets import load_iris
iris=load_iris()
```

```python
dir(iris)
```

```python
df=pd.DataFrame(iris.data,columns=iris.feature_names)
df.head()
df["target"]=iris.target
df
```

```python
df["flower_name"]=df.target.apply(lambda x:iris.target_names[x]) ################make  a new colum###############################
df
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
df0= df[df.target==0]
df1=  df[df.target==1]
df2= df[df.target==2]
```

```python
plt.xlabel('petal width (cm)')
plt.ylabel('petal length (cm)')
plt.scatter(df0['petal length (cm)'],df0['petal width (cm)'], color="green",marker="+")
plt.scatter(df1['petal length (cm)'],df1['petal width (cm)'],color="green",marker="+")
```

```python
x0=df.drop(["target","flower_name"], axis="columns")
x0
```

```python
y0=df.target
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x0,y0,test_size=0.2)
```

```python
from sklearn.svm import SVC
model=SVC()

model.fit(x_train,y_train)
# len(x_train),len(y_train)
```

```python
model.predict(x_test)
model.score(x_test,y_test) ###########0.96 or 1 ######################################################
```
