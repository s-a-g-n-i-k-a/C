NAIVE BAYES (01)

```python
import pandas as pd
df=pd.read_csv("titanic.csv")
df
df.columns
```

```python
df1=df.drop(['PassengerId','Name','SibSp','Parch','Ticket','Cabin','Embarked'],axis='columns')
df1
```

```python
target=df1['Survived']
target
```

```python
inputs=df1.drop(["Survived"],axis="columns")
inputs
```

```python
dummies=pd.get_dummies(inputs.Sex, dtype=int)
dummies
```

```python
input=pd.concat([inputs,dummies],axis="columns")
input
```

```python
input.columns[input.isna().any()]#######vvimp
input.Age=inputs.Age.fillna(inputs.Age.median())

input=input.drop(['Sex'], axis="columns")
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(input,target,test_size=0.2)
```

```python
from sklearn.naive_bayes import GaussianNB
model= GaussianNB()
model.fit(x_train,y_train)
model.score(x_test,y_test)
```
